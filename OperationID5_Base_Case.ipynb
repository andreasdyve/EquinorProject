```python
#KASPER
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_excel('op5.xlsx')

df1 = df.copy()
```

```python
x = df1['timeVectRelMinutes']
y1 = df1['sumEngMdoPower']
y2 = df1['thrusterAndPropulsion']
y5 = df1['battery_0__actualPower']

#x = df1.iloc[72639:92505]['timeVectRelMinutes']
#y1 = df1.iloc[72639:92505]['Battery_Out']
#y2 = df1.iloc[72639:92505]['thrusterAndPropulsion']
#y3 = df1.iloc[72639:92505]['sumEngMdoPower']
#y4 = df1.iloc[72639:92505]['Battery_In']
#y5 = df1.iloc[72639:92505]['battery_0__actualPower']

fig = plt.figure()

plt.plot(x, y1, label = 'Engine Power')
plt.plot(x, y2, label = 'Propulsion Power')
plt.plot(x, y5, label = 'Battery Power')
plt.legend(loc="upper center", bbox_to_anchor=(0.5, 1.15), ncol=2)
plt.xlabel('Time')
plt.ylabel('Output')

plt.rcParams['figure.figsize'] = [12, 8]
plt.rcParams['figure.dpi'] = 100

plt.draw()

# Update the plot
plt.show()
```
