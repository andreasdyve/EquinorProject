```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('data_full_sekres.xlsx')
```

```python
df['engMdoConsumption_0_cum_'] = df['engMdoConsumption_0_'].cumsum()/(60*60)
df['engMdoConsumption_1_cum_'] = df['engMdoConsumption_1_'].cumsum()/(60*60)
df['engMdoConsumption_2_cum_'] = df['engMdoConsumption_2_'].cumsum()/(60*60)
df['engMdoConsumption_3_cum_'] = df['engMdoConsumption_3_'].cumsum()/(60*60)

df['engMdoConsumption_all_cum_'] = df['engMdoConsumption_0_cum_'] + df['engMdoConsumption_1_cum_'] +  \
df['engMdoConsumption_2_cum_'] + df['engMdoConsumption_3_cum_']

df['engMdoConsumption_all_cum_'] = df['engMdoConsumption_all_cum_']

df.plot(x='datetime', y=['engMdoConsumption_0_cum_', 'engMdoConsumption_1_cum_', \
                         'engMdoConsumption_2_cum_', 'engMdoConsumption_3_cum_'])
plt.xlabel('time')
plt.ylabel('liters')

df.plot(x='datetime', y='engMdoConsumption_all_cum_')
plt.xlabel('time')
plt.ylabel('liters')
```
