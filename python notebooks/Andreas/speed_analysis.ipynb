```python
import pandas as pd
df = pd.read_excel('dataTableStudents_I_2.xlsx')
```

```python
df2 = pd.read_excel('dataTableStudents_II_2.xlsx')
```

```python
df2
```

```python
df3 = df.join(df2.drop(columns='timeVectRelMinutes'))
df3
```

```python
df_op3 = df3.loc[df3['operationId'] == 3]
df_op3.plot(x='timeVectRelMinutes', y='speedOverGround')

df_op5 = df3.loc[df3['operationId'] == 5]
df_op5.plot(x='timeVectRelMinutes', y='speedOverGround')
```

```python
df3.rename(columns= {'timeVectRelMinutes': 'datetime'}, inplace=True)
df3['datetime'] = pd.to_datetime(df3['datetime'], unit='m', origin='2020-01-30')
minute_df = df3.resample('min', on='datetime').mean().reset_index()
```

```python
min_df_op3 = minute_df.loc[minute_df['operationId'] == 3]
min_df_op3.plot(x='datetime', y='speedOverGround')

min_df_op5 = minute_df.loc[minute_df['operationId'] == 5]
min_df_op5.plot(x='datetime', y='speedOverGround')
```
