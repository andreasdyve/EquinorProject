# Importerer pandas og leser inn filene

```python
import pandas as pd
df = pd.read_excel('dataTableStudents_I.xlsx')
df2 = pd.read_excel('dataTableStudents_II.xlsx')
```

```python
df3 = df.join(df2.drop(columns='timeVectRelMinutes'))
df3
```

# Endrer navn på og gjør om tid-kolonnen til datetime

## datetime er en standardisert måte å vise tid på, trengs for å bruke funksjonen resample() til å aggregere

```python
df3.rename(columns= {'timeVectRelMinutes': 'datetime'}, inplace=True)
df3['datetime'] = pd.to_datetime(df3['datetime'], unit='m', origin='2020-01-30')
```

# Resampler til minutt- og timesoppløsning

```python
min_df = df3.resample('min', on='datetime').mean().reset_index()
hour_df = df3.resample('60min', on='datetime').mean().reset_index()
```

# Lagrer filene som nye excel-filer

```python
min_df.to_excel('data_full_minres.xlsx')
hour_df.to_excel('data_full_hourres.xlsx')
df3.to_excel('data_full_sekres.xlsx')
```
