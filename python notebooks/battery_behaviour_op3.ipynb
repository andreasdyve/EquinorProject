```python
import pandas as pd
df = pd.read_excel('op3.xlsx')
```

```python
df
```

```python
df.rename(columns= {'timeVectRelMinutes': 'datetime'}, inplace=True)
df['datetime'] = pd.to_datetime(df['datetime'], unit='m', origin='2020-01-30')
```

```python
df['index'] = df.index
```

```python
import matplotlib.pyplot as plt

df.loc[38730:38850].plot(x='datetime', y='thrusterAndPropulsion', figsize=(11, 7), color='red')
plt.xlabel('timestamp')
plt.ylabel('output[kW]')
plt.savefig('peakshaving_t&p.png')

df.loc[38730:38850].plot(x='datetime', y=['engMdoPower_0_', 'engMdoPower_1_', 'engMdoPower_3_',\
                                     'thrusterAndPropulsion'], figsize=(11, 7))
plt.xlabel('timestamp')
plt.ylabel('output[kW]')
plt.savefig('peakshaving_t&p_dg_battery.png')

df.loc[38730:38850].plot(x='datetime', y='battery_0__actualPower', figsize=(11, 7))
plt.xlabel('timestamp')
plt.ylabel('output[kW]')
plt.savefig('peakshaving_battery.png')
```
