```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('data_full_sekres.xlsx')
```

```python
df_4 = df.copy()

#Create column of battery output and set negatives as zero
df_4['Battery_Out'] = df_4['battery_0__actualPower']
num = df_4['Battery_Out']._get_numeric_data()
num[num < 0] = 0

#Create column of battery input and set positives as zero
df_4['Battery_In'] = df_4['battery_0__actualPower']
num = df_4['Battery_In']._get_numeric_data()
num[num > 0] = 0

#Normalize battery output
df_4['Battery_Out'] = (df_4['Battery_Out'] - df_4['Battery_Out'].min()) / (df_4['Battery_Out'].max() - df_4['Battery_Out'].min())

#Normalize sum power output
df_4['sumEngMdoPower'] = (df_4['sumEngMdoPower'] - df_4['sumEngMdoPower'].min()) / (df_4['sumEngMdoPower'].max() - df_4['sumEngMdoPower'].min())

df_4.plot(x = 'datetime', y = ['Battery_Out', 'sumEngMdoPower'], title = 'Time Series Power Output MB and Diesel Generators')
plt.xlabel('Time')
plt.ylabel('Output')
```

```python
#Now to see when the generators charge the battery
#First Set Power in as positive
df_4['Battery_In'] = df_4['Battery_In']*(-1)

#Second Normalize
df_4['Battery_In'] = (df_4['Battery_In'] - df_4['Battery_In'].min()) / (df_4['Battery_In'].max() - df_4['Battery_In'].min())

#Third plot
df_4.plot(x = 'datetime', y = ['Battery_In', 'sumEngMdoPower'], title = 'Time Series MB Charge and Diesel Generator Output')
plt.xlabel('Time')
plt.ylabel('Output')

display(df_4)
```

```python
df_4.plot(x = 'datetime', y = ['Battery_In', 'Battery_Out'])
```
